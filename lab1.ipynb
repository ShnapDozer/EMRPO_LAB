```python
import numpy as np
from scipy.special import factorial
from scipy.optimize import bisect

def f_log(n):
    return np.log(n)

def f_sqrt(n):
    return np.sqrt(n)

def f_linear(n):
    return n

def f_square(n):
    return n**2

def f_cube(n):
    return n**3

def f_exponential(n):
    return 2**n

def f_factorial(n):
    return factorial(n, exact=False)

def solve_log(t):
    return 2**t

def solve_sqrt(t):
    return t**2

def solve_linear(t):
    return t

def solve_square(t):
    return np.sqrt(t)

def solve_cube(t):
    return t**(1/3)

def solve_exponential(t):
    return np.log2(t)

def solve_factorial(t):
    return bisect(lambda n: factorial(n) - t, 1, 1000)
```

```python
import pandas as pd
from tabulate import tabulate

time_units = {
      "секунда":  1.0
    , "минута":   60.0
    , "час":      3600.0
    , "день":     24.0 * 3600
    , "месяц":    30.0 * 24 * 3600
    , "год":      365.0 * 24 * 3600
    , "век":      100.0 * 365 * 24 * 3600
}

functions = {
      "lg n":     solve_log
    , "sqrt(n)":  solve_sqrt
    , "n":        solve_linear
    , "n^2":      solve_square
    , "n^3":      solve_cube
    , "2^n":      solve_exponential
    , "n!":       solve_factorial
}

results = {key: [] for key in functions}

for time_label, time_value in time_units.items():
    for func_name, solver in functions.items():
        try:
            n_max = solver(time_value)
            results[func_name].append(n_max)
        except:
            results[func_name].append(float('nan'))

table = pd.DataFrame(results, index=time_units.keys())

print(tabulate(table, headers='keys', tablefmt='psql'))
```

```python
import matplotlib.pyplot as plt

x = np.linspace(1, 100, 1000)
plt.figure(figsize=(12, 8))
# plt.yscale("log")
plt.ylim(0, 30)
plt.xlim(0, 30)

plt.plot(x, f_log(x), label="lg n")
plt.plot(x, f_sqrt(x), label="sqrt(n)")
plt.plot(x, f_linear(x), label="n")
plt.plot(x, f_square(x), label="n^2")
plt.plot(x, f_cube(x), label="n^3")
plt.plot(x, f_exponential(x), label="2^n", linestyle="--")
plt.plot(x[x < 20], f_factorial(x[x < 20]), label="n!", linestyle=":")  # Ограничение для factorial

plt.legend()
plt.xlabel("n")
plt.ylabel("f(n)")
plt.title("Графики функций f(n)")
plt.grid()
plt.show()
```
